```python
class MyCircularQueue():
#source: https://www.programiz.com/dsa/circular-queue
    def __init__(self, k):
        self.k = k
        self.queue = [None] * k
        self.head = self.tail = -1
#----------------------------------------
    def enqueue(self, data):

        if ((self.tail + 1) % self.k == self.head):
            print("The circular queue is full\n")

        elif (self.head == -1):
            self.head = 0
            self.tail = 0
            self.queue[self.tail] = data
        else:
            self.tail = (self.tail + 1) % self.k
            self.queue[self.tail] = data
#----------------------------------------------
    def isEmpty(self):
        if self.tail == -1:
            return bool(True)
        return bool(False)
#----------------------------------------    
    def isFull(self):
        if ((self.tail + 1) % self.k == self.head):
            return bool(True)
        return bool(False)
#----------------------------------------    
    def dequeue(self):
        if (self.head == -1):
            print("The circular queue is empty\n")

        elif (self.head == self.tail):
            temp = self.queue[self.head]
            self.head = -1
            self.tail = -1
            return temp
        else:
            temp = self.queue[self.head]
            self.head = (self.head + 1) % self.k
            return temp
#----------------------------------------------
    def getFront(self):
        if (self.head == -1):
            print("The circular queue is empty\n")
        else:
            return self.queue[self.tail]
#--------------------------------------------        
    def printCQueue(self):
        if(self.head == -1):
            print("No element in the circular queue")

        elif (self.tail >= self.head):
            for i in range(self.head, self.tail + 1):
                print(self.queue[i], end=" ")
            print()
        else:
            for i in range(self.head, self.k):
                print(self.queue[i], end=" ")
            for i in range(0, self.tail + 1):
                print(self.queue[i], end=" ")
            print()
```

```python
#question 6:

class Queue:
    def __init__(self):
        self.items = []
        self.size = 0
        
    def enqueue(self, data):
        self.size += 1
        self.items.append(data)
 
    def dequeue(self):
        self.size -= 1
        return self.items.pop(0)
    
    def is_empty(self):
        return self.items == []
    
    def get_size(self):
        return self.size
    
    def displayQueue(self):
        if self.size>0:
            print(self.items)


class Stack:
    def __init__(self):
        self.Q = Queue()
    
    def push(self, data):
        self.Q.enqueue(data)
    
    def pop(self):
        if not self.Q.is_empty():
            for _ in range(self.Q.get_size() - 1):
                dequeued = self.Q.dequeue()
                self.Q.enqueue(dequeued)
            return self.Q.dequeue()
        else:
            return "queue is empty"
    def is_empty(self):
        return self.Q.is_empty()
    
    def displayStack(self):
        if not (self.Q.is_empty()):
            self.Q.displayQueue()
    
#main:
list=[]
stak=Stack()
size = 5
for i in range(size):
    stak.push(i)

stak.displayStack()
for i in range(0 , size):
    list.append(stak.pop())
print(list)
```
